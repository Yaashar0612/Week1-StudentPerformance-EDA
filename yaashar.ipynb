```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df.isnull().sum()
```

```python
print("Average scores by gender:\n", df.groupby('gender')[['math score', 'reading score', 'writing score']].mean())
print("\nAverage scores by parental level of education:\n", df.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean())
print("\nAverage scores by lunch type:\n", df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean())
print("\nAverage scores by test preparation course:\n", df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='math score', data=df)
plt.title('Average Math Score by Gender')
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Score Distribution by Test Preparation')
plt.show()
plt.figure(figsize=(6,4))
sns.histplot(df['reading score'], kde=True, color='orange')
plt.title('Distribution of Reading Scores')
plt.show()
plt.figure(figsize=(6,4))
sns.boxplot(x='lunch', y='writing score', data=df)
plt.title('Writing Score Distribution by Lunch Type')
plt.show()

```

```python
print("""
Key Insights:

1. On average, female students scored higher in reading and writing, while male students scored slightly higher in math.
2. Students whose parents have higher education levels tend to perform better across all subjects.
3. Students with standard lunch performed better than those with free/reduced lunch.
4. Students who completed the test preparation course scored significantly higher in math, reading, and writing.
""")
```
